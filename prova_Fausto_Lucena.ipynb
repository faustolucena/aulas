```python
nota = float(input('Digite a nota da avaliação: '))
lista = []
while nota!=-1:
  if nota>=0:
    lista.append(nota)
  elif nota<-1:
    print(f"O número {nota} é invalido. Digite outro número ou -1 para sair.")
  if nota == -1:
    break
  nota = float(input('Digite a nota da avaliação: '))  
#obtem maximo e minimo
if len(lista)>0:
  max = lista[0]
  min = lista[0]
  for x in lista:
    if max<x:
      max=x
    if min>x:
      min=x
  #saída  
  print(f"Os valores digitados foram {lista}.")
  print(f"O maior valor é {max}")
  print(f"O menor valor é {min}")
else: #verifica se houve valores de entrada
  print(f"Não houve valores de entrada.")
```

```python
def myfunction(string1,string2):
  #string1 = palavra que deve ser contida
  #string2 = frase que devemos verificar
  s1 = string1.upper()
  s2 = string2.upper()
  remover = [".", ",", ":", ";", "!", "?"]
  for caracter in remover:
    s2 = s2.replace(caracter, " ")
  
  mysplit = s2.split()
  flag = 0
  for x in mysplit:
      if s1 == x:
        flag = 1
  if flag == 1:
    print('Entrada:')
    print(f'  - String1 = {string1}')
    print(f'  - String2 = {string2}')
    print("Saída: True")
  else:
    print('Entrada:')
    print(f'  - String1 = {string1}')
    print(f'  - String2 = {string2}')
    print("Saída: False") 

```

```python
myfunction("olá","Olá, João! Tudo bem?")
```

```python
myfunction("água","Gostaria de tomar um café agora.")
```

```python
def meuspospares(lista):
  pares = []
  for x in lista:
    if (x%2)==0:
      pares.append(x)
  print("Entrada:")
  print(f"  - {lista}")
  print("Saída:")
  print(f"  - {pares}")
```

```python
#Carrega os meus 10 números
n = 1
lista = []
while n<=10:
  numero = int(input(f'Digite o {n} número inteiro positivo de {10}: '))
  if numero<=0:
    print(f'O número {numero} não é um inteiro positivo')
  else:
    n+=1
    lista.append(numero)

#executa minha função
meuspospares(lista)
```
