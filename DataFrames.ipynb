```python
import pandas as pd
import numpy as np
url =('https://raw.githubusercontent.com/jennybc/'
      'gapminder/master/data-raw/08_gap-every-five-years.tsv')
dfPaises = pd.read_csv(url, sep='\t')
dfPaises.shape
```

```python
dfPaises.head()
```

```python
dfPaises.rename(columns={'country':'Pais',
                         'continent':'Continente',
                         'year':'Ano',
                         'lifeExp':'ExpVida',
                         'pop':'População',
                         'gdpPercap':'PibPerCap'}, inplace = True)
dfPaises.head()
```

```python
dfPaises.columns
```

```python
dfPaises = dfPaises[['Continente', 'Pais', 'Ano','ExpVida','População','PibPerCap']]
```

```python
dfPaises.head()
```

```python
dfPaises.info()
```
